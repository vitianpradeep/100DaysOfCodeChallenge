```python
import matplotlib.pyplot as plt
import pandas as pd

iris = pd.read_csv('D:\MCA\II SEM\ML\Assignment\iris.csv')
print(iris.head())

y = iris[['SepalLengthCm']]
x = iris[['PetalLengthCm']]
x.columns = ['PetalLengthCm']

from sklearn.model_selection import train_test_split

#training model and also breaking into 70:30 ratio
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)

from sklearn.linear_model import LinearRegression
#Fitting linear model
lr = LinearRegression()
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)
print(y_pred)
y_test.head(), y_pred[0:5]

from sklearn.metrics import mean_squared_error
print("mean_squared_error is : ",mean_squared_error(y_test,y_pred))
print("Slope is : ",lr.coef_)
print("Intercept is : ",lr.intercept_)
#prediction of sample

#Plotting the data set and regression model
plt.scatter(x, y)
plt.plot(x, lr.predict(pd.DataFrame({'PetalLengthCm': x['PetalLengthCm']})), color='blue', label='Regression Line')
plt.legend()
plt.show()
```
