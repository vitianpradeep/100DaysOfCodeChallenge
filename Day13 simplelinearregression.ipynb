# Linear Regression example

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

## Step 2: Load the CSV file

```python
data = pd.read_csv('/kaggle/input/salary-data-with-age-and-experience/Salary_Data.csv')
```

## Step 3: Explore the dataset

```python
print(data.head())  # to view the first few rows of the data
print(data.describe())  # to view the statistical summary of the data
```

## Step 4: Prepare the data for modeling

### For simple linear regression, we need to select one feature (also called independent variable or predictor variable) and one target variable (also called dependent variable). For multiple linear regression, we can select multiple features and one target variable.For the purpose of this example, let's assume we have a CSV file with two columns 'experience' and 'salary', and we want to perform linear regression to predict the salary based on the experience. We will use the 'experience' column as the feature and 'salary' column as the target.

```python
# Simple linear regression
X = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values.reshape(-1, 1)

# # Multiple linear regression
# X = data[['YearsExperience', 'education', 'age']]
# y = data['salary'].values.reshape(-1, 1)
```

## Step 5: Split the data into training and testing sets

```python
x_test,x_train, y_test,y_train = train_test_split(X,y, test_size = 0.2)
```

## Step 6: Create and train the linear regression model

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

## Step 7: Make predictions and evaluate the model

```python
# Predict the salary for the test data
y_pred = model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)

# Evaluate the model using mean squared error
print("mean squared error is : ", mse)
```

## Step 8: Visualize the data and the model

```python
#simple linear Regression
plt.scatter(X,y)
plt.plot(X,model.predict(X), color='red')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```
